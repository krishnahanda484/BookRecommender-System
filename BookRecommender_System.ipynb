##Colaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
experienced_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(experienced_users)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
import numpy as np

def recommend(book_name):
    book_name = book_name.strip().lower()  # Normalize input

    # Check if book exists in pt.index
    book_titles = [title.strip().lower() for title in pt.index.astype(str)]  # Normalize dataset

    if book_name not in book_titles:
        return f"Error: '{book_name}' not found. Please enter a valid book title."

    # Get book index safely
    index = book_titles.index(book_name)

    # Fetch similar items
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]

    data = []
    for i in similar_items:
        temp_df = books[books["Book-Title"].str.lower() == pt.index[i[0]].strip().lower()]

        if temp_df.empty:
            continue

        item = [
            temp_df["Book-Title"].values[0],
            temp_df["Book-Author"].values[0],
            temp_df["Image-URL-M"].values[0]
        ]
        data.append(item)

    if not data:
        return "Error: No similar books found."

    return data
```
